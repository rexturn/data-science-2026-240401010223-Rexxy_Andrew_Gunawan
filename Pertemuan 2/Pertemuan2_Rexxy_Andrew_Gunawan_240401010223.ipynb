# 3. Filtering

## Penumpang yang selamat

```python
# Penumpang yang selamat adalah yang memiliki nilai df['Survived'] == 1
penumpang_selamat = df[df['Survived']==1]
print(f"Jumlah penumpang yang selamat : {penumpang_selamat.shape[0]} orang")
print(f"Persentase penumpang yang selamat : {penumpang_selamat.shape[0]/df.shape[0]*100} %")
```

##  Penumpang Wanita Kelas 1

```python
wk1 = df[(df['Sex']=='female') & (df['Pclass']==1)] # filter penumpang wanita & di Kelas 1
print(f"Jumlah penumpang wanita di kelas 1 adalah : {wk1.shape[0]} orang")

wk1_survived = wk1[wk1['Survived']==1] # penumpang wanita di Kelas 1 'yang selamat'
print(f"Jumlah penumpang wanita yang selamat di Kelas 1 : {wk1_survived.shape[0]} orang")

print(f"Rata-rata penumpang wanita yang selamat di Kelas 1 : {wk1['Survived'].mean()*100} %")

print(f"Rata-rata usia penumpang kelas 1 adalah : {wk1['Age'].mean()} tahun")

print(f"Rata-rata usia penumpang wanita yang selamat di Kelas 1 adalah {wk1_survived['Age'].mean()} tahun")
```

# 4. Groupby - Tingkat Keselamatan per Kelas


```python
hasil = df.groupby('Pclass').agg(
  total=('PassengerId','count'),
  selamat=('Survived','mean'),
).round(3)
print(hasil)
```

## Q5
Gunakan groupby untuk menghitung tingkat keselamatan(mean Survived) per kelas penumpang. Kelas mana yang paling tinggi?

Jawab:

Kelas yang memiliki tinggkat keselamatan paling tinggi adalah "Kelas 1".

```python
# mencari variabel lain (jenis kelamin) - untuk Q6
hasil = df.groupby('Sex').agg(
  total=('PassengerId','count'),
  selamat=('Survived','mean'),
)
print(hasil)
```
